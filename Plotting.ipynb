```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("/content/imdb_top_1000.csv")
```

```python
genre_distrbution = data_frame["Genre"].value_counts()
```

```python
plt.figure(figsize=(30,10))
genre_distrbution.plot(kind ="bar")
plt.ylabel("Number of Movies")
plt.xlabel("Genre Type")
plt.title("Genre Distrbution")
plt.show()
```

```python
movies_per_year = data_frame.groupby('Released_Year').size()
plt.figure(figsize=(30,10))
plt.scatter(movies_per_year.index, movies_per_year.values)
plt.ylabel("Number of Movies")
plt.xlabel("Year")
plt.title("Number of Movies per Year")
plt.show()
```
